df.nsmallest(n, columns, keep='first') -> 'DataFrame'
Docstring:
Return the first `n` rows ordered by `columns` in ascending order.

Return the first `n` rows with the smallest values in `columns`, in
ascending order. The columns that are not specified are returned as
well, but not used for ordering.
ITS similar to sort_values(),but better.

```python
#for bottom half of income earners,
len(df)/2
```

```python
#lets create a col for these bottom halfs,order/SORT by amt_income_total col.(n is int,dont use decimals)
bottom_half_income= df.nsmallest(n=219278,columns='AMT_INCOME_TOTAL')
```

```python
sns.boxplot(data=bottom_half_income,x='NAME_FAMILY_STATUS',y='AMT_INCOME_TOTAL',hue='FLAG_OWN_REALTY')
```

```python
plt.figure(figsize=(10,4),dpi=100)
sns.boxplot(data=bottom_half_income,x='NAME_FAMILY_STATUS',y='AMT_INCOME_TOTAL',hue='FLAG_OWN_REALTY')
```

```python
plt.figure(figsize=(10,4),dpi=100)
sns.boxplot(data=bottom_half_income,x='NAME_FAMILY_STATUS',y='AMT_INCOME_TOTAL',hue='FLAG_OWN_REALTY') 
#order=['Married','Widow','Civil marriage''Separated'])
plt.title('Income totals per family status for bottom half of earners')


plt.legend(loc=(1.05,0),title='FLAG_OWN_REALTY')
```
